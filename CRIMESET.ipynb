```python
import pandas as pd
from uszipcode import SearchEngine
search = SearchEngine(simple_zipcode=True)
from uszipcode import Zipcode
import numpy as np
```

```python
data=pd.read_csv(r'C:\sarah school\New folder\SPD.csv')
```

```python
data
```

```python
#This function wil check all the columns and return the number of rows with NaN values
def check_null_col(col):
    null_columns=col.columns[col.isnull().any()]
    return col[null_columns].isnull().sum()

dataset_check_NaN=check_null_col(data)
print("Name of column and number of Null values in the dataset")
print(dataset_check_NaN)
print("Size of dataset =",data.shape)
```

```python
#Removing all the rows with NaN values to avoid discrepency
dataset_temp=data
dataset_temp=dataset_temp.dropna()
dataset_temp=dataset_temp.reset_index(drop=True)
print(dataset_temp.shape)
dataset_temp
```

```python
#Here we will seperate date and time into seperate cells for easier operations
datetime = pd.to_datetime(dataset_temp["Offense Start DateTime"]) 
dataset_temp['timestamp']=datetime
dataset_temp['Date'] = dataset_temp['timestamp'].dt.date
dataset_temp['Year'] = dataset_temp['timestamp'].dt.year
dataset_temp['Month_num'] = dataset_temp['timestamp'].dt.month
dataset_temp['Time'] = dataset_temp['timestamp'].dt.time
```

```python
dataset_temp
```

```python
dataset_temp = dataset_temp[dataset_temp.Year == 2020]
```

```python
dataset_temp
```

```python
#Renaming columns
dataset_temp.rename(columns={'Offense Code':'offense_code'},inplace=True)
dataset_temp.rename(columns={'Offense':'offense'},inplace=True)
dataset_temp.rename(columns={'Precinct':'precinct'},inplace=True)
dataset_temp.rename(columns={'Sector':'sector'},inplace=True)
dataset_temp.rename(columns={'Beat':'beat'},inplace=True)
dataset_temp.rename(columns={'100 Block Address':'address'},inplace=True)
dataset_temp.rename(columns={'Longitude':'longitude'},inplace=True)
dataset_temp.rename(columns={'Latitude':'latitude'},inplace=True)
dataset_temp.rename(columns={'Date':'date'},inplace=True)
dataset_temp.rename(columns={'Year':'year'},inplace=True)
dataset_temp.rename(columns={'Time':'time'},inplace=True)
dataset_temp.rename(columns={'Month':'month'},inplace=True)
dataset_temp.rename(columns={'Report Number':'report_number'},inplace=True)
dataset_temp.rename(columns={'Offense ID':'offense_id'},inplace=True)
dataset_temp.rename(columns={'Offense Start DateTime':'offense_start_time'},inplace=True)
dataset_temp.rename(columns={'Offense End DateTime':'offense_end_time'},inplace=True)
dataset_temp.rename(columns={'Report DateTime':'report_time'},inplace=True)
dataset_temp.rename(columns={'Group A B':'group'},inplace=True)
dataset_temp.rename(columns={'Crime Against Category':'crime_category'},inplace=True)
dataset_temp.rename(columns={'Offense Parent Group':'offense_group'},inplace=True)

#Deleting unnecessary columns
dataset_temp.drop("timestamp",axis=1,inplace=True,errors='ignore')
```

```python
dataset_temp
```

```python
#Function to get zipcode from lat and long
def get_zipcode(lat, lon):
    result = search.by_coordinates(lat = lat, lng = lon, returns = 1)
    return result[0].zipcode
```

```python
dataset_temp['zipcode']=dataset_temp.apply(lambda x: get_zipcode(x.latitude,x.longitude),axis=1)
```

```python
dataset_temp
```

```python
total_cases=dataset_temp.groupby(['zipcode']).count()
total_cases=total_cases.loc[:,['report_number']]
total_cases.rename(columns={'report_number':'cases_reported'},inplace=True)
total_cases
```
